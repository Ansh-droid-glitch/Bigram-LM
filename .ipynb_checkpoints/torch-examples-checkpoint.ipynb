```python
import numpy as np
import time
import torch
device = 'cpu'
print(device)
```

```python
%%time
start_time = time.time()
#Matrix operations here
zeros = torch.zeros(1, 1)
end_time = time.time()

elapsed_time = end_time - start_time
print(f"Elapsed time:{elapsed_time: .8f}")
```

```python
torch_rand1 = torch.rand(100, 100, 100, 100).to(device)
torch_rand2 = torch.rand(100, 100, 100, 100).to(device)
np_rand1 = torch.rand(100, 100, 100, 100)
np_rand2 = torch.rand(100, 100, 100, 100)

start_time = time.time()

rand = (torch_rand1 @ torch_rand2)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time:.8f} sec")

start_time = time.time()

rand = np.multiply(np_rand1, np_rand2)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time:.8f} sec")
#gpu faster than cpu for bigger tasks
```

```python
import torch.nn as nn
sample = torch.tensor([10.,10.,10.])
linear = nn.Linear(3, 3, bias=False)
print(linear(sample))
```

```python
import torch.nn.functional as F

#Create a tensor
tensor1 = torch.tensor([1.0, 2.0, 3.0])  

#Apply softmax using torch.nn.functional.softmax()
softmax_output = F.softmax(tensor1, dim=0)

print(softmax_output)
```

```python
#Initialize an embedding layer
vocab_size = 1000
embedding_dim = 100
embedding = nn.Embedding(vocab_size, embedding_dim)

#create some imput indicies
input_indicies = torch.LongTensor([1, 5, 3, 2])

#Apply the embedding layer
embedded_output = embedding(input_indicies)

#The output will be a tesnor of (4, 100), where 4 is the number of inputs
# and 100 is the dimensity of the embedding vectors
print(embedded_output.shape)
```

```python
a = torch.tensor([[1, 2], [3, 4], [5, 6]])
b = torch.tensor([[7,8,9],[10,11,12]])
print(a @ b)
print(torch.matmul(a, b))
```

```python
int_64 = torch.randint(1, (3, 2))
#type int64
float_32 = torch.rand(2, 3)
#type float32
result = torch.matmul(int_64, float_32)
print(result)
#we cant multiple other dtypes with others
```

```python
#but we can change the int to float
int_64 = torch.randint(1, (3, 2)).float()
#type int64
float_32 = torch.rand(2, 3)
#type float32
result = torch.matmul(int_64, float_32)
print(result)
```
